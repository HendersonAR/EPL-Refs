```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import cm
```

```python
data = pd.read_csv("/Users/juan/Downloads/JAMAMA.csv")
#You need to change this

home = 3
away = 4
hbp = 27
abp = 28
hf = 16
af = 17 

```

```python
def readit(file):
    lst = []
    with open(file, "r") as myfile:
        csvreader = csv.reader(myfile)
        for row in csvreader:
            lst.append(row)
        lst.pop(0)
    return lst

# Read data
data = readit("/Users/juan/Downloads/JAMAMA.csv")

data[0]

def teams(lst):
    equipes = []
    for row in lst:
        if row[home] not in equipes:
            equipes.append(row[home])
    return equipes
x = teams(data)
x = sorted(x)
x

    
```

```python
def bpavg(lst, team):
    total = 0
    count = 0
    for row in lst:
        if row[home] == team:
            total += int(row[hbp])
            count +=1
        elif row[away] == team:
            total += int(row[abp])
            count +=1
        else:
            continue
    avg = total / count
    return round(avg, 2)
```

```python
def foulsavg(lst, team):
    total = 0
    count = 0
    for row in lst:
        if row[home] == team:
            total += int(row[hf])
            count +=1
        elif row[away] == team:
            total += int(row[af])
            count +=1
        else:
            continue
    avg = total / count
    return round(avg, 2)
```

```python
def bpteams():
    equipes = {}
    for team in x:
        equipes[team] = bpavg(data, team)
    return equipes

def foulsby():
    fouls = {}
    for team in x:
        fouls[team] = foulsavg(data, team)
    return fouls
    
y = bpteams()
uhh = foulsby()
y
```

```python
average_bp = list(y.values())
average_fls = list(uhh.values())
```

```python
def normalize(values):
    normd = []
    maxi = max(values)
    mini = min(values)
    denomi = maxi - mini
    for val in values:
        newnew = (val - mini) / denomi
        normd.append(newnew)
    return normd
```

```python
norm_bp = normalize(average_bp)
norm_fls = normalize(average_fls)
norm_bpf = normalize(bp_per_foul)
```

```python
# Plotting the lines
plt.figure(figsize=(12, 6))

# Plot for Average Booking Points
plt.plot(x, norm_bp, marker="o", linestyle="-", color="#D62728", linewidth=2, label="Normalized Avg Booking Points")

# Plot for Average Fouls
plt.plot(x, norm_fls, marker="s", linestyle="--", color="#2CA02C", linewidth=2, label="Normalized Avg Fouls")

# Plot for Booking Points per Foul
# plt.plot(x, norm_bpf, marker="^", linestyle=":", color="#1F77B4", linewidth=2, label="Normalized Booking Points per Foul")

# Adding title and labels
plt.title("Team Average Booking Points, Fouls, and BP per Foul", fontsize=16)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Rotate x-axis labels for clarity
plt.xticks(rotation=45, ha='right')

# Add a grid for better readability
plt.grid(alpha=0.7)

# Add a legend to differentiate lines
plt.legend(fontsize=12)

# Adjust layout for spacing
plt.tight_layout()

# Show the plot
plt.show()
```
